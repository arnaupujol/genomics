```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from genomics_modules import barcodes as bc
from genomics_modules import stats, plotting, utils
import datetime
import pdb
```

### Setting file names, paths and parameters  

```python
data_path = '/Users/arnau/Documents/isglobal/genomics/data/'
file_name = data_path + 'Genotyping_results_March2020_1234-PF-MZ-MAYOR.xlsx'
filename_wgs_sample = data_path + 'Mayor_sample_selection.csv'
```

```python
data_path = '/Users/arnau/Documents/isglobal/genomics/data/DB_ARNAU/'
filename_mda2_pcr = data_path + 'mda2_pcr.csv'
filename_codes_WT_ISG = data_path + '0_Equiv_codigos_WT_ISGlobal_19March2020.dta'
filename_MipMon = data_path + '1_MiPMon_19March2020.dta'
filename_MDA1 = data_path + '2_MDA1_19March2020.dta'
filename_MDA2 = data_path + '3_MDA2_19March2020.dta'
filename_REACT_contact = data_path + '4_REACT_contact_19March2020.dta'
filename_REACT_index = data_path + '5_REACT_index_19March2020.dta'
filename_Crosses = data_path + '6_Crosses_19March2020.dta'
filename_geoMDA2 = data_path + '7_geoMDA2_19March2020.dta'
filename_geoReactMDA1 = data_path + '8_geoReactMDA1_19March2020.dta'
filename_geoMiPMon = data_path + '9_geoMiPMon_19March2020.dta'
```

### Loading data  

```python
grc = pd.read_excel(file_name)
grc2 = pd.read_excel(file_name, sheet_name = 1)
barcodes = pd.read_excel(file_name, sheet_name = 2)
grc['Date']= pd.to_datetime(grc['Date of Collection'])
```

```python
mda2_pcr = pd.read_csv(filename_mda2_pcr)
codes_WT_ISG = pd.read_stata(filename_codes_WT_ISG)
MipMon = pd.read_stata(filename_MipMon)
MDA1 = pd.read_stata(filename_MDA1)
MDA2 = pd.read_stata(filename_MDA2)
REACT_contact = pd.read_stata(filename_REACT_contact)
REACT_index = pd.read_stata(filename_REACT_index)
Crosses = pd.read_stata(filename_Crosses)
geoMDA2 = pd.read_stata(filename_geoMDA2)
geoReactMDA1 = pd.read_stata(filename_geoReactMDA1)
geoMiPMon = pd.read_stata(filename_geoMiPMon)
```

```python
wgs_sample = pd.read_csv(filename_wgs_sample)
```

```python
#Merging genotypes with nida in new dataframe
grc_nida = pd.merge(grc, codes_WT_ISG, left_on = 'Sample External ID', right_on = 'external_id')
```

```python
#Merging info of whether the sample is selected for WGS
grc_nida = pd.merge(grc_nida, wgs_sample, left_on = 'Sample Internal ID', right_on = 'Sample_ID', how = 'left')
```

### Making one data frame with all the interesing data integrated

```python
#Simplifying grc table to interested columns
grc_all = grc_nida[['external_id', 'nida', 'Location', 'Species', 'COIL', 'Barcode', 'Date', 'estudi', 'Selected_for_sequencing']]
#Unifying MipMon data
MipMon_all = pd.merge(MipMon[['nida', 'visit', 'centro', 'visdate', 'permid', 'age', 'gestnum']], geoMiPMon[geoMDA2['nida'].notnull()], on = 'nida', how = 'left')
grc_all = pd.merge(grc_all, MipMon_all[['nida', 'visit', 'centro', 'visdate', 'permid', 'age', 'gestnum', 'longitude', 'latitude']], on = 'nida', how = 'left')
#Unifying all MDA data
MDA_all = pd.merge(MDA1[['nida', 'subdistrict', 'age', 'gender']], MDA2[['nida', 'permid', 'date']][np.isfinite(MDA2['nida'])], on = 'nida', how = 'left')
geoMDA2['nida'][geoMDA2['nida'] == ''] = 'NaN'
geoMDA2['nida'] = geoMDA2['nida'].astype(float)
MDA_all = pd.merge(MDA_all, geoMDA2[np.isfinite(geoMDA2['nida'])], on = 'nida', how = 'left')

grc_all = pd.merge(grc_all, MDA_all[['nida', 'subdistrict', 'age', 'gender', 'permid', 'date', 'longitude', 'latitude']], on = 'nida', how = 'left', suffixes = ('_Mip', '_MDA'))
grc_all = grc_all.rename(columns = {'gender' : 'gender_MDA', 'date' : 'date_MDA', 'visdate' : 'visdate_Mip'})

REACT_contact = REACT_contact.drop_duplicates()
grc_all = pd.merge(grc_all, REACT_contact[['nida', 'permid', 'age', 'gender', 'is_pregnant']], on = 'nida', how = 'left')
grc_all = grc_all.rename(columns = {'permid' : 'permid_react', 'age' : 'age_react', 'gender' : 'gender_react'})

REACT_index = REACT_index.drop_duplicates()
REACT_index['nida'] = REACT_index['nida'].astype(float)
grc_all = pd.merge(grc_all, REACT_index[['nida', 'date', 'permid', 'gender', 'birthday', 'is_pregnant']], on = 'nida', how = 'left')
grc_all = grc_all.rename(columns = {'date' : 'date_react', 'permid' : 'permid_react2', 'gender' : 'gender_react2'})

grc_all = pd.merge(grc_all, Crosses[['nida', 'studyarea', 'permid', 'lng', 'lat', 'area', 'visdate', 'age', 'gender', 'pregnancy']][np.isfinite(Crosses['nida'])], on = 'nida', how = 'left')
grc_all = grc_all.rename(columns = {'permid' : 'permid_cross', 'visdate' : 'visdate_cross', 'age' : 'age_cross', 'gender' : 'gender_cross'})

grc_all = pd.merge(grc_all, geoReactMDA1[['nida', 'longitude', 'latitude']][np.isfinite(geoReactMDA1['nida'])], on = 'nida', how = 'left')
grc_all = grc_all.rename(columns = {'longitude' : 'longitude_react', 'latitude' : 'latitude_react'})
```

```python
grc_all = grc_all.drop_duplicates()
```

### Basic numbers of the data  

```python
valids = bc.barcodes_are_valid(grc_all['Barcode'])
failed = np.array([i == '-' for i in grc_all['Barcode']], dtype=bool)
monoclonal = bc.barcodes_are_monoclonal(grc_all['Barcode'])
polyclonal = bc.barcodes_are_polyclonal(grc_all['Barcode'])

print("Number of samples: " + str(len(grc_all['Barcode'])))
print("Number of bases in barcodes: " + str(len(grc_all['Barcode'][0])))
print("Number of failed barcodes: " + str(np.sum(failed)))
print("Number of valid barcodes: " + str(np.sum(valids)))
print("Number of unique barcodes: " + str(len(grc_all['Barcode'][valids].unique())))
print("Number of monoclonal barcodes: " + str(np.sum(monoclonal)))
print("Number of polyclonal barcodes: " + str(np.sum(polyclonal)))
print("Number of unique monoclonal barcodes: " + str(len(grc_all['Barcode'][monoclonal].unique())))
print("Number of unique polyclonal barcodes: " + str(len(grc_all['Barcode'][polyclonal].unique())))
```

```python
locs = []
counts = []
for loc in grc['Location'].unique():
    locs.append(loc)
    counts.append(np.sum(valids*(grc_all['Location'] == loc)))
plt.figure(figsize=[6,6])
plt.pie(counts, labels = locs, autopct='%1.1f%%')
plt.show()
```

```python
are_in_mipmon = grc_all['age_Mip'].notnull()
total_mipmon = np.sum(are_in_mipmon)
print("Number of samples from MipMon: " + str(total_mipmon))
are_in_MDA = grc_all['age_MDA'].notnull()
total_MDA = np.sum(are_in_MDA)
print("Number of samples from MDA: " + str(total_MDA))
are_in_react = grc_all['is_pregnant_x'].notnull() + grc_all['is_pregnant_y'].notnull()
total_react = np.sum(are_in_react)
print("Number of samples from REACT: " + str(total_react))
are_in_cross = grc_all['visdate_cross'].notnull()
total_cross = np.sum(are_in_cross)
print("Number of samples from Cross: " + str(total_cross))
print("Total: " + str(np.sum(are_in_mipmon) + np.sum(are_in_MDA) + np.sum(are_in_react) + np.sum(are_in_cross)))
```

```python
plt.figure(figsize=[6,6])
plt.pie([total_mipmon, total_MDA, total_react, total_cross], labels = ['MipMon', 'MDA', 'REACT', 'Crosses'], autopct='%1.1f%%')
plt.show()
```

### Analysing polyclonal cases, heterozygous calls and missed bases 

```python
count_n = np.array([bar.count('N') for bar in grc['Barcode'][polyclonal]])
count_x_poly = np.array([bar.count('X') for bar in grc['Barcode'][polyclonal]])

count_x_mono = np.array([bar.count('X') for bar in grc['Barcode'][monoclonal]])
```

# Biallelicness of barcodes

Here I find how many different bases can be found in each SNPs.   

```python
#color code of bases
base2color = {
    'A' : 'tab:blue',
    'C' : 'tab:orange',
    'G' : 'tab:green',
    'T' : 'tab:red',
    'X' : 'w',
    'N' : 'tab:gray',
    '-' : 'w',
    'm' : 'lightgray'
}
```

```python
monoclonal_bars = np.array([bc.barcode_list(bar) for bar in grc['Barcode'][monoclonal]])
unique_bases_mono = np.array([bc.count_unique_bases(monoclonal_bars[:,i]) for i in range(monoclonal_bars.shape[1])])
polyclonal_bars = np.array([bc.barcode_list(bar) for bar in grc['Barcode'][polyclonal]])
unique_bases_poly = np.array([bc.count_unique_bases(polyclonal_bars[:,i]) for i in range(polyclonal_bars.shape[1])])
```

```python
plt.plot(unique_bases_mono, alpha = .5, label = 'monoclonal')
plt.plot(unique_bases_poly, alpha = .5, label = 'polyclonal')
plt.title('Number of different bases per SNPs on the samples')
plt.xlabel('SNP number')
plt.ylabel('Number of unique bases')
plt.legend(frameon = False)
plt.show()
```

I found one SNP to have up to three bases for monoclonal and polyclonal samples. It corresponds to SNP 39

```python
np.unique(monoclonal_bars[:,39]), np.unique(polyclonal_bars[:,39])
```
