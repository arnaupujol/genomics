```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Setting file names, paths and parameters 

```python
data_path = '/Users/arnau/Documents/isglobal/genomics/data/DB_ARNAU/'
filename_mda2_pcr = data_path + 'mda2_pcr.csv'
filename_codes_WT_ISG = data_path + '0_Equiv_codigos_WT_ISGlobal_19March2020.dta'
filename_MipMon = data_path + '1_MiPMon_19March2020.dta'
filename_MDA1 = data_path + '2_MDA1_19March2020.dta'
filename_MDA2 = data_path + '3_MDA2_19March2020.dta'
filename_REACT_contact = data_path + '4_REACT_contact_19March2020.dta'
filename_REACT_index = data_path + '5_REACT_index_19March2020.dta'
filename_Crosses = data_path + '6_Crosses_19March2020.dta'
filename_geoMDA2 = data_path + '7_geoMDA2_19March2020.dta'
filename_geoReactMDA1 = data_path + '8_geoReactMDA1_19March2020.dta'
filename_geoMiPMon = data_path + '9_geoMiPMon_19March2020.dta'
```

### Loading data frame from excel file  

```python
mda2_pcr = pd.read_csv(filename_mda2_pcr)
codes_WT_ISG = pd.read_stata(filename_codes_WT_ISG)
MipMon = pd.read_stata(filename_MipMon)
MDA1 = pd.read_stata(filename_MDA1)
MDA2 = pd.read_stata(filename_MDA2)
REACT_contact = pd.read_stata(filename_REACT_contact)
REACT_index = pd.read_stata(filename_REACT_index)
Crosses = pd.read_stata(filename_Crosses)
geoMDA2 = pd.read_stata(filename_geoMDA2)
geoReactMDA1 = pd.read_stata(filename_geoReactMDA1)
geoMiPMon = pd.read_stata(filename_geoMiPMon)
```

```python
geo_MDA1 = pd.merge(geoReactMDA1, MDA1, on = 'nida', how = 'inner')
```
