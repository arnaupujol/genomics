# Example barcode scripts 

This notebook shows some example scripts to use and visualise barcodes from the genomic_tools repository. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from genomic_tools import barcodes, plotting
```

The module barcodes includes methods to analyse barcodes. Let's assume the following barcodes:

```python
barcode_1 = 'ATCGCATAGACACCCTAGAAAAGTTCTC'
barcode_2 = 'ATATCAAAGCCACCATATGAAAATCCTC'
```

We can measure the fraction of common bases between both as: 

```python
difference_fraction = barcodes.diff_count(barcode_1, barcode_2)
print('Fraction of common bases: ', difference_fraction)
```

And check what bases are equal: 

```python
#base-to-base comparison (True if equal, False if different)
barcodes.bar_diff(barcode_1, barcode_2)
```

 We can obtain a relatedness matrix with all the pair comparisons for a list of barcodes. 

```python
barcode_3 = 'AATGCAAATGATCATTATAAAGCCCTTC'
barcode_4 = 'ATTGAAAATATACCAAACGAGTATGCCA'
```

```python
relatedness_mat = barcodes.relatedness_matrix([barcode_1, barcode_2, barcode_3, barcode_4])
```

```python
plt.imshow(relatedness_mat)
plt.xlabel('Sample 1')
plt.ylabel('Sample 2')
plt.title('Fraction of common bases')
plt.colorbar()
```

And other simple operations, such as identifying specific bases in the barcode. For example, find where the barcode has 'A':

```python
#To identify when do we find a given base
barcodes.base_is(barcode_1, 'A')
```

### Plotting 

```python
#Dictionary defining a colour per base
base2color = {
    'A' : 'tab:red',
    'T' : 'tab:orange',
    'C' : 'tab:blue',
    'G' : 'tab:green',
}
```

We can plot a colour coded barcode, or a list of barcodes: 

```python
plotting.plot_barcode(barcode_1, base2color, ysize = 1, s = 100)
```

```python
plotting.plot_barcodes([barcode_1, barcode_2, barcode_3, barcode_4], base2color, s = 100)
```

We can also plot the differences with respect to a reference, showing in green the matched bases 

```python
base2color = {
    'A' : 'tab:red',
    'T' : 'tab:orange',
    'C' : 'tab:blue',
    'G' : 'tab:green',
    'm' : 'tab:grey',
}
```

```python
plotting.plot_bar_diff(barcode_1, [barcode_2, barcode_3, barcode_4], base2color)
```
